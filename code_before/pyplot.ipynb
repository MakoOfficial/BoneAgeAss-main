```python
data = np.random.rand(10,2)  # 生成数据
idx = [i for i in range(3, 13)]  # 可以设置索引
df = pd.DataFrame(data, columns=['列1', '列2'], index=idx)
# df = pd.DataFrame(data, columns=['列1', '列2'])   # 使用默认索引
ax = df.plot(kind='bar')
ax.set_xlabel('年份')  # 设置x轴标签
ax.set_ylabel('分数') # 设置y轴标签
fig = ax.get_figure()  # 用于保存图片
fig.savefig('pic.png')  # 保存为png格式
```

```python
Id, age, gender = myKit.read_labels('./dataset/', 'boneage-training-dataset.csv')
record = torch.zeros(230)
for i in range(age.shape[0]):
    record[int(age[i])] += 1
idx = [i for i in range(0, 230)]
df = pd.DataFrame(record, columns=['数量'], index=idx)
ax = df.plot(kind='bar')
ax.set_xlabel('年龄')
ax.set_ylabel('人数')
fig = ax.get_figure()
```
